This notebook objective is to perform an exploratory data analysis on the Student Performance dataset available at UCI Machine Learning repository. This analysis follows the steps bellow:

1. Understand the data instances and attributes
2. Understand the distribution of each attribute
3. Understand the joint distribution of attributes
4. Understand correlation between variables
5. Understand how the independent attributes can be predict the dependent variable

```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

import warnings
warnings.filterwarnings("ignore")

%matplotlib inline
```

## Loading the data

Originally, the dataset is compressed in a .zip file, so we downloaded to our notebook directory. The dataset can be found at https://archive.ics.uci.edu/ml/datasets/Student+Performance.

```python
students = pd.read_csv('data/student-mat.csv', sep=';')
students.head()
```

```python
# grades distribution

plt.figure(figsize=(20, 10))
plt.subplot(2, 3, 1)
sns.distplot(students.G1)
plt.xlim(0, 20)
plt.subplot(2, 3, 4)
sns.boxplot(students.G1)

plt.subplot(2, 3, 2)
sns.distplot(students.G2)
plt.xlim(0, 20)
plt.subplot(2, 3, 5)
sns.boxplot(students.G2)


plt.subplot(2, 3, 3)
sns.distplot(students.G3)
plt.xlim(0, 20)
plt.subplot(2, 3, 6)
_ = sns.boxplot(students.G3)
```
